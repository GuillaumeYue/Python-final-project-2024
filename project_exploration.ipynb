```python
import requests from IPython.display import display, Image import random # Function to get a random Pokémon ID (there are 898 Pokémon as of Generation 8) def get_random_pokemon_id():    return random.randint(1, 898) # Fetch and display 10 random Pokémon images for _ in range(10):    pokemon_id = get_random_pokemon_id()    url = f"https://raw.githubusercontent.com/PokeAPI/sprites/master/sprites/pokemon/{pokemon_id}.png"    response = requests.get(url)    if response.status_code == 200:        display(Image(response.content))    else:        print(f"Failed to fetch Pokémon image for ID: {pokemon_id}")
```
